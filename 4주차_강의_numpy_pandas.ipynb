# numpy 기본 기능
numpy는 수치해석, 통계 관련 기능을 구현하는데 기본이 되는 함수이다.

```python
import numpy as np
print(np.sqrt(2))
```

```python
import numpy as np
print(np.power(2,4))
print(np.power(3,3))
```

```python
import numpy as np
print(np.pi)
print(np.sin(0))
print(np.cos(np.pi))
```

```python
#랜덤 함수
import numpy as np
a=np.random.rand(5) #0~1 사이 랜덤 실수 5개
print(a)
```

```python
print(np.random.choice(6,10)) # 0~5 사이 정수 10개 선택
```

```python
print(np.random.choice(10,6, replace=False))  #이미 뽑은 숫자 안 뽑기
```

```python
print(np.random.choice(6,10,p=[0.1,0.2,0.3,0.2,0.1,0.1])) #p로 0~5 숫자 나올 확률 지정
```

# 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np
dice=np.random.choice(6,10)

plt.hist(dice,bins=6)
plt.show()
```

```python
#위에서 지정한 확률 그래프로 확인해보기
histogram=np.random.choice(6,1000000,p=[0.1,0.2,0.3,0.2,0.1,0.1])
plt.hist(histogram, bins=6)
plt.show()


plt.savefig('myfigure.png', dpi=300)  #그림을 해상도 300으로 저장
```

# 판다스 기본기능
pandas는 numpy를 기반으로 만들어졌다.
1차원 배열 형태는 series, 2차원 배열 형태의 구조는 dataframe이라고 부른다.
dataframe을 다루는 방법을 알아보자.


*   준비사항: 구글에서 '올림픽 메달 집계' 검색 후 위키페이지의 주소를 복사한다



```python
import pandas as pd
df=pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
print(df)
```

```python
df[0]  #df리스트의 0번 인덱스에 저장된 내용 보기
```

```python
import pandas as pd
df=pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84', header=0, index_col=0) #맨 위의 헤더를 열 이름, 나라 이름을 인덱스 이름으로
df[0]
```

```python
#하계올림픽만 추출하기

summer=df[0].iloc[:,:5]  #iloc-> 명시된 숫자만큼 슬라이싱, 0~5번째 열만 summer라는 데이터프레임에 저장
summer
```

```python
#열 이름 지정
summer.columns=['하계참가횟수', '금', '은', '동', '계']
summer
```

```python
#금메달 갯수대로 정렬하기
print(summer.sort_values('금', ascending=False))
```

```python
#표를 엑셀로 저장
summer=summer.sort_values('금', ascending=False)
summer.to_excel('summer_olympics.xlsx')
```

```python
summer[['금']].var() # 표본 분산
```

```python
summer[['금']].sum() # 합계
```

```python
summer[['금']].max() # 최대값
```

```python
summer[['금']].quantile(0.75) # 제 2 사분위수
```

# 시각화 연습

```python
#히스토그램 그리기
import matplotlib.pyplot as plt
import pandas as pd

plt.hist(summer['금'], bins=10)
plt.show()
```

```python
#박스플롯 그리기
import matplotlib.pyplot as plt
import pandas as pd

plt.boxplot(summer['금'])
plt.show()
```

```python
#박스플롯 그리기
import matplotlib.pyplot as plt
import pandas as pd

plt.boxplot(summer['은'])
plt.show()

#이상치 때문에 그림이 유용하지 못함
```

```python
#이상치 제거
summer2 = summer[summer['은'] < 1000]
summer2
```

```python
#박스플롯 그리기
import matplotlib.pyplot as plt
import pandas as pd

plt.boxplot(summer2['은'])
plt.show()
```

```python
#이상치 제거
summer3 = summer[summer['은'] < 30]
summer3
```

```python
#박스플롯 그리기
import matplotlib.pyplot as plt
import pandas as pd

plt.boxplot(summer3['은'])
plt.show()
```

```python
#금과 은 함께 그리기
import matplotlib.pyplot as plt
import pandas as pd

plt.boxplot([summer3['금'],summer3['은']])
plt.show()
```

```python
# 두 개 이상의 조건 --> & 로
summer4 = summer[(summer['은'] < 30) & (summer['은'] > 5)]

plt.boxplot(summer4['은'])
plt.show()
```

```python
#라벨 붙이기
import matplotlib.pyplot as plt
import pandas as pd

plt.boxplot([summer3['금'], summer3['은']], tick_labels=['Gold', 'Silver'])  # labels 추가
plt.show()
```
